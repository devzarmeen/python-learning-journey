How to deal with missing values in EDA Analysis 
categorical/object , Numeric , Boolean ,

1. Numerical Data(Int,float)
Replace different values of numerical variable  
it is float number we can impute it with 3 to 4 methods
where missing value is in age we will replace it with average of age  OR mean
Method 1: Fill with Mean (Best for normally distributed data)
Method 2: Fill with Median   (Recommended for Skewed Data)
Method 3: Fill with Constant (0,-1)
Method 4: Interpolation


2. Categorical Data
Method 1 :Replace it with mode of age  
Method 2 :Fill or replace with unknown
Method 3 :Forward fill(ffill)
Method 4 :Backward fill(bfill)


3. Boolean Data

Method 1:Replace missing values with true  

Method 2:Replace missing values with false

Method 3:Replace missing values with mode 

4. Forward Fill (ffill)

Missing values filled from prevous row 

Synatx: df.ffill()

5. Backward Fill (Bfill)

Missing values Filled from next row

SYNATX: df.bfill()

6. Drop Missing Values 

if we have many missing values then remove only rows 

SYNTAX: df.dropna()

7. Fill Entire DataFrame 

Fill numeric columns only 

SYNTAX: df = df.fillna(df.mean(numeric_only=True))

we canot replace categorical data  if we do we will get incorrect answers and we have to do extra calculations
 

```python
## FILL NUMERIC
df["Age"] = df["Age"].fillna(df["Age"].median())

df["Fare"] = df["Fare"].fillna(df["Fare"].mean())
```

```python
## FILL CATEGORICAL 
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])

df["Sex"] = df["Sex"].fillna(df["Sex"].mode()[0])
```

## Mean, Median, Mode

Mean: The average. Add all numbers together, then divide by the total count

Median: The middle number. Order the numbers from smallest to largest and pick the exact middle
2,3,5,7 -> center 3,5 -> average of center -> 3+5/2 = 4 -> median = 4

Mode: The most frequent number. The number that appears the most times in a data set
2,3,5,6,6,9,10,9 -> mode -> (6,9) -> (first mode, second mode) -> (index 0 , index 1)
in code -> mode()[0] -> [0] -> means first mode which is 6
