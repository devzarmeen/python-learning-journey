2. Displot (Distribution Plot )

It explain distribution of data 

```python
import matplotlib.pyplot as plt ##  Display graph 
import seaborn as sns   ## import seaborn
sns.displot([0,1,2,3,4,5])  ## COnvert list into graph 
plt.show()   ## Will show graph 
```

KDE

```python
import matplotlib.pyplot as plt ##  Display graph 
import seaborn as sns   ## import seaborn
sns.displot([0,1,2,3,4,5],kind="kde")  # kde means Draw smooth curve not histrogram 
plt.show()  ## show plot

# DIsttribution means how data is spread 
```

3. Normal Distribution

It is Guassian distribution

IQ, Height, Weight, ExamMarks, Blood Pressure, Heartbeat

It forms Bell Curve

```python
## random.normal()
## random.normal(loc,scale,size) it has 3 parameters 
## loc  explain mean, Average, Bell Curve Center
## scale  (Standard Deviation) means how data is spreaded
## size   how many random numbers generated 
## size(2,3) 2 rows and 3 columns 
from numpy import random
x=random.normal(size=(2,3))
print(x)
```
