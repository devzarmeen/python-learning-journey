## 1.Numpy Summation 

What is difference between summation and Addition?

Addition add two values and summation is in between multiple values 

Mathematics Notation: ∑

```python
## Example 
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
newarr = np.add(arr1, arr2)
print("Addition")
print(newarr)

print("")
print("Sum")
## Example 
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
newarr = np.sum([arr1, arr2])
print(newarr)

print("")
print("Sum with vertices ")
## Example 
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([1,2,3])
newarr = np.sum([arr1, arr2],axis=1)
print(newarr)
## axis=0 = vertically (columns)
## axis=1 = horizontally (rows)


## Cummulative sum (cumsum)
print("")
print("Cumulative Sum")
import numpy as np
arr=np.array([1,2,3])
newarr=np.cumsum(arr)
print(newarr)
```

## 2. NUMPY Production 

Multiply Numbers 

```python
## Example 1
import numpy as np
arr = np.array([1,2,3,4])
x = np.prod(arr)
print("Product of single array")
print(x)

print("")
## Example 2 
import numpy as np
arr1 = np.array([1,2,3,4])
arr2 = np.array([5,6,7,8])
x = np.prod([arr1,arr2])
print("Product of two arrays")
print(x)

print("")
## Product over axis       it gives row by row answer
import numpy as np
arr1=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
newarr=np.prod([arr1,arr2],axis=1)
print("Production over axis")
print(newarr)

print("")
## Cumulative Product   it save product of each state and each state answer multiplies with next value
import numpy as np
arr=np.array([5,6,7,8])
newarr=np.cumprod(arr)
print("Cumulative Product")
print(newarr)
```

## 3.NUMPY Difference

Calculate Difference of two consecutive values 

Next Element - Previous Element

or 

Current Value - Previous Value

```python
import numpy as np
arr = np.array([10, 15, 25, 5])   ## when next value is small then we have negative value in return
newarr = np.diff(arr)
print("Difference of an array")
print(newarr)

print("")
import numpy as np
arr = np.array([10, 15, 25, 5])
newarr = np.diff(arr, n=2)
print("Differnec Opernation twice")
print(newarr)
```

## 4.NumPy LCM Lowest Common Multiple

The Lowest Common Multiple is the smallest number that is a common multiple of two numbers.

```python
import numpy as np
num1 = 4
num2 = 6
x = np.lcm(num1, num2)
print("LCM")
print(x)

print("")
## LCM of Array 
import numpy as np
arr = np.array([3, 6, 9])
x = np.lcm.reduce(arr)   ## reduce() apply same function on each element of array repeadtely 
print("LCM of array")
print(x)

print("")
## Arrange 
import numpy as np
arr = np.arange(1,11)   ## arrange indictes start and stop value
x=np.lcm.reduce(arr)
print("LCM")
print(x)
```

## 5.NumPy GCD Greatest Common Divisor

The GCD (Greatest Common Divisor), also known as HCF (Highest Common Factor) is the biggest number that is a common factor of both of the numbers.

```python
import numpy as np
num1 = 6
num2 = 9
x = np.gcd(num1, num2)
print("GCD")
print(x)

print("")
print("GCD in Array")
import numpy as np
arr = np.array([20, 8, 32, 36, 16])
x = np.gcd.reduce(arr)
print(x)
```

## 6.NUMPY TRIGNOMETRIC FUNCTIONS

NumPy provides the ufuncs sin(), cos() and tan() that take values in radians and produce the corresponding sin, cos and tan values.

```python
import numpy as np
x = np.sin(np.pi/2)
print("The Sine value of pI/2")
print(x)

## Find sine values for all of the values in arr:
print("")
import numpy as np
arr = np.array([np.pi/2, np.pi/3, np.pi/4, np.pi/5])
x = np.sin(arr)
print("Sine values for all of the values in arr:")
print(x)

## Convert Degrees Into Radians
print("")
import numpy as np
arr = np.array([90, 180, 270, 360])
x = np.deg2rad(arr)
print("Convert Degrees Into Radians")
print(x)

##Radians to Degrees
print("")
import numpy as np
arr = np.array([np.pi/2, np.pi, 1.5*np.pi, 2*np.pi])
x = np.rad2deg(arr)
print("Convert Radians to Degrees")
print(x)

## Finding Angles 
print("")
## arcsin()
import numpy as np
x = np.arcsin(1.0)
print("Finding Angle using arcsin()")
print(x)

print("")
import numpy as np
arr = np.array([1, -1, 0.1])
x = np.arcsin(arr)
print("Finding Angle of each value using arcsin()")
print(x)

## Hypotenues
##  takes the base and perpendicular values and produces hypotenues based on pythagoras theorem.
import numpy as np
base = 3
perp = 4
x = np.hypot(base, perp)
print("Hypotenues for base 4 and perpendicular 3")
print(x)

```

## 7. NUMPY Hyperbolic Functions

## 8.NUMPY Set Operations 

What is set 

A set in mathematics is a collection of unique elements.

Sets are used for operations involving frequent intersection, union and difference operations.

```python
import numpy as np
arr = np.array([1, 1, 1, 2, 3, 4, 5, 5, 6, 7])
x = np.unique(arr)
print("Create Unique set")
print(x)


## Finding Union   To find the unique values of two arrays
print("")
print("Finding Union")
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
newarr = np.union1d(arr1, arr2)
print(newarr)
```

```python
## Findng  Intersection
## To find only the values that are present in both arrays
print("")
import numpy as np
arr1 = np.array([1, 2, 3, 4])
arr2 = np.array([3, 4, 5, 6])
newarr = np.intersect1d(arr1, arr2, assume_unique=True)
print("Intersection of two arrays")
print(newarr)

## Finding DIfference 
print("")
import numpy as np
set1 = np.array([1, 2, 3, 4])
set2 = np.array([3, 4, 5, 6])
newarr = np.setdiff1d(set1, set2, assume_unique=True)
print("The difference of the set1 from set2:")
print(newarr)

## Finding Symmetric Difference
## To find only the values that are NOT present in BOTH sets
print("")
import numpy as np
set1 = np.array([1, 2, 3, 4])
set2 = np.array([3, 4, 5, 6])
newarr = np.setxor1d(set1, set2, assume_unique=True)
print("Symmetric Difference of two sets")
print(newarr)
```
