## Python Datatypes 

```python
# Python Data Types Examples
# 1. String (str)
name = "Alisha"              #  Text, always written inside quotes (" " or ' ')

# 2. Integer (int)
age = 20                  # Whole numbers (no decimal)

# 3. Float (float)
price = 99.5              #Decimal numbers

# 4. Complex (complex)
x = 5 + 3j                #  Numbers with a real and imaginary part used in mathematical calculations 

# 5. List (list)
fruits = ["Apple", "Banana", "Cherry"]   # Ordered, mutable (can be changed), uses [ ]

# 6. Tuple (tuple)
colors = ("Red", "Green", "Blue")        # Ordered, immutable (cannot be changed), uses ( )

# 7. Range (range)
numbers = range(5)        # Range = Sequence of numbers, mostly used in loops

# 8. Dictionary (dict)
student = {
    "name": "Ali",
    "age": 20
}                          # Stores data as key:value pairs, uses { }

# 9. Set (set)
unique_fruits = {"Apple", "Banana", "Cherry"}   # Unordered collection of unique values

# 10. FrozenSet (frozenset)
fixed_set = frozenset({"Apple", "Banana"})      # Immutable version of a set(not changeable)

# 11. Boolean (bool)
is_student = True         #Only two values: True or False

# 12. Bytes (bytes)
data = b"Hello"           # Immutable binary data, used in files and networking

# 13. Bytearray (bytearray)
byte_data = bytearray(5)  # Mutable binary data

# 14. MemoryView (memoryview)
memory = memoryview(bytes(5))   # MemoryView = Efficient way to access binary data without copying

# 15. NoneType (None)
value = None              #Represents no value or empty value


# type() Function
type(name)                # Returns the data type of the variable
type(age)
type(price)

# Constructors (Specific Data Types)

text = str("Hello")                     # Create a String
number = int(20)                        # Create an Integer
decimal = float(20)                     # Create a Float
fruit_list = list(("Apple", "Banana"))  # Create a List
fruit_tuple = tuple(("Apple", "Banana"))# Create a Tuple
fruit_set = set(("Apple", "Banana"))    # Create a Set
person = dict(name="Alisha", age=20)       # Create a Dictionary
status = bool(5)                        # Create a Boolean (Non-zero values become True)
```

## Membership Operator
Used to check whether a value exists in a collection (List, String, Tuple, Set, or Dictionary).

in        Checks if the value exists in the collection.

not in    Checks if the value does not exist in the collection.

```python
fruits = ["apple", "banana", "cherry"]

print("banana" in fruits)

fruits = ["apple", "banana", "cherry"]

print("pineapple" not in fruits)
```

```python
# membership operator with strings
text = "Hello World"
print("h" in text) ## due to case sensitivity it will show false
```

```python
## Membership in Tuple
text = "Hello World"
print("H" in text)

#Membership in Set
colors = {"red", "green", "blue"}
print("yellow" in colors)

#Membership in Dictionary
student = {
    "name": "Ali",
    "age": 20
}

print("name" in student)
```
