2. Pandas- Cleaning Empty Cell 

## Empty Cell 

Empty cells can potentially give us a wrong result when we analyze data.

Generally Empty cells denoted by NAN (Not A Number)

## Remove Rows

One way to deal with empty cells is to remove rows that contain empty cells.

This is usually OK, since data sets can be very big, and removing a few rows will not have a big impact on the result.

## dropna()
By default, the dropna() method returns a new DataFrame, and will not change the original.

Return a new Data Frame with no empty cells:

```python
import pandas as pd
df = pd.read_csv(r"E:\Internship\practiceai\Pandas_Cleaning.csv")
new_df = df.dropna()
print(new_df.to_string()) ## to_string() print complete dataframe
```

## inplace = True
If you want to change the original DataFrame, use the inplace = True argument:

Remove all rows with NULL values

```python
import pandas as pd

df = pd.read_csv(r"E:\Internship\practiceai\Pandas_Cleaning.csv")

df.fillna(0, inplace=True)

print(df.to_string())
```

## dropna(inplace = True)

Now, the dropna(inplace = True) will NOT return a new DataFrame, but it will remove all rows containing NULL values from the original DataFrame.

Permanently delete

```python
import pandas as pd

df = pd.read_csv(r"E:\Internship\practiceai\Pandas_Cleaning.csv")

df.dropna(inplace = True)

print(df.to_string())
```

## Replace only specified column 

fillna()

Instead of deleting, replace missing values


```python
import pandas as pd

df = pd.read_csv(r"E:\Internship\practiceai\Pandas_Cleaning.csv")
df.fillna({"Calories":130}, inplace=True)
print(df.to_string())
```

## Replace with mean median mode

```python
x = df["Calories"].mean()
print(x)

df.fillna({"Calories":x}, inplace=True)
df
```

## median 

```python
x= df['Calories'].median()
print(x)

df.fillna({"Calories":x}, inplace=True)
df
```

## Mode 

```python
x = df["Calories"].mode()[0]
print(x)

df.fillna({"Calories":x}, inplace=True)
df
```

```python
for x in df.index:
    if df.loc[x,"Duration"]>120:
        df.drop(x,inplace=True)
```
