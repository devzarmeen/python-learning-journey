1. Random Number

Random number does NOT mean a different number every time. Random means something that can not be predicted logically.
Each time output will be different

True Random vs Pseudo Random

True Random
* Completely unpredictable.
* Comes from a source outside the computer.
* **Examples:**

  * Mouse movement
  * Keyboard typing
  * Network timing
* Mostly used for security, such as encryption.

Pseudo Random

* Generated by a computer algorithm.
* It is not truly random, but it appears random to us.
* NumPy uses pseudo-random numbers.



```python
## Import Random Module

from numpy import random

## random.randint()  it generate random integer
x = random.randint(100)
print(x)

print("")
## range
print("Show Range")
random.randint(50,100)
print(x)

## random.rand()    generate float 
print("")
print("show decimal random")
from numpy import random
x=random.rand()
print(x)

## random Array 
print("")
print("Show Random Array")
x=random.randint(100,size=(5))
print(x)
## 2D Array
print("")
print("show 2D Array")
x=random.randint(100,size=(3,5))
print(x)

## Random Flat Array
print("")
print("Random Float ")
x=random.rand(5)
print(x)

## random.choice()
print("")
print("Random Choice")
x=random.choice([3,5,7,9])
print(x)
```

3. Random Permutation

Change Arrangements

```python
## Shuffle()               change the original Array
from numpy import random
import numpy as np
arr=np.array([1,2,3,4,5])
random.shuffle(arr)
print("Shuffle() Method")
print(arr)

## Permutation() Method     original array remains same
print("")
print("Permutation() Method")  
arr=np.array([1,2,3,4,5])
x=random.permutation(arr)
print(x)
print(arr)
```
