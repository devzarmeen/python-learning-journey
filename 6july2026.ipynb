## 1. Python _ Range
It is a built-in Python function that generates a sequence of numbers.

```python
print("Range")
for i in range(5):  # This is the stop value. The start value is included, but the stop value is not included.
    print(i)

print("")    
print("range using start and stop")
for i in range(3,8):
    print(i)

print("")    
print("range using start stop and step")
for i in range(2,11,2):
    print(i)
    
## Negative Steps
print("")    
print("range using start stop and negativestep")
for i in range(10,0,-1):
    print(i)

## Range Object
print("") 
print("Range Object")
x = range(5)  # its not list its an object 
print(list(x))
print("Length of range")
print(len(x))

## List Range
print("") 
print("List Range ")
print(list(range(5)))


## for loop in Range 
print("")
print("for Loop in Range")
for i in range(4):
    print("Hello")   # range tells loop number of iteration not start stoop step values
    
## Index of string in range 
print("")
print("Index of string in range ")
name = "Ali"
for i in range(len(name)):
    print(i)
    
## Slicing in Range 
print("")
print("Slicing in range")
r=range(10)
print(r[2:8])
print("list of slicing range")
print(list(r[2 : 8]))

## Membership (IN)
print("")
print("Membership (IN)")
r = range(0,10,2)
print(6 in r)
```

## Python _ Array 
python doesnot have built in array just like c and c++ and java thats why we use list as a array 

```python
print("Array in Python and index")
cars = ["Ford", "BMW", "Honda", "Toyota", "Audi"]
print(cars[1])
print(cars[4])
print("")
print("Negative index")
print(cars[-2])
print(cars[-3])

## Change value in array
print("")
print("Change the value in list  or array")
cars = ["Ford","Volvo","BMW"]
cars[0] = "Toyota"
print(cars)

## length of array 
print("")
print("Length of array elements ")
cars=["ford","BMW","Volvo"]
print(len(cars))

## Loop in array 
print("")
print("loop in array")
cars = ["Ford", "BMW", "Honda", "Toyota", "Audi"]
for x in cars:
    print(x)
    
## Append Method in array
print("")
print("Append Method in array")
cars=["ford","BMW","Volvo"]
cars.append("Honda")
print(cars)

## pop method in array
print("")
print("pop() Method in array")
cars=["ford","BMW","Volvo"]
cars.pop(0)
print(cars)

## remove() method in array
print("")
print("remove() Method in array")
cars = ["Ford","Volvo","BMW"]
cars.remove("Volvo")
print(cars)

## clear() method in array 
print("")
print("clear() Method in array")
cars = ["Ford","Volvo","BMW"]
cars.clear()
print(cars)

## copy() Method  in array 
print("")
print("copy() Method in array")
cars = ["Ford","Volvo","BMW"]
newcars = cars.copy()
print(newcars)

## count() method in array 
print("count()) Method in array")
numbers = [1,2,3,2,2,5]
print(numbers.count(2)) ## it will count number of 2's in this list

## insert() method in array
## it insert at specific position 
print("")
print("insert() Method in array")
cars = ["Ford","BMW"]
cars.insert(1,"Volvo")
print(cars)

## extend() method
print("") ## combine 2 lists
print("extend() method in array")
cars = ["Ford","BMW"]
more = ["Honda","Toyota"]
cars.extend(more)
print(cars)

## Reverse() Method
print("") ## combine 2 lists
print("reverse() method in array")
cars = ["Ford","BMW","Honda"]
cars.reverse()
print(cars)
```

## 3. Python _ Iterator
iterator is an object that return values one by one
1. iterable         object that can be loop(list,tuple,set ,dict,string)   we can execute loop on these 
2. iterator         object  that gives next values

```python
## iter() method   it is used  to form iterator
print("Form iterator using iter() method")
fruits = ("apple", "banana", "cherry")
myit = iter(fruits) ## A fruit iterator (object) is created, which returns the next value, and it is stored in `myit`.
print(myit)

## next() method   ## Used to get the next value from an iterator.
print("")
print("next value using next() method from iter method()")
fruits = ("apple", "banana", "cherry")
myit = iter(fruits)
print(next(myit))
print(next(myit))
print(next(myit))

## __iter__() and __next__() Method  Advance
print(" __iter__() and __next__() Method  Advance ")
## iter() Method 
print("__iter__ Method")
class MyNumbers:
    def __iter__(self):
        self.a = 1
        return self
## __next__() Method 
    print("__next__() Method ")
    def __next__(self):
        if self.a<=5:
         x = self.a
         self.a += 1
         return x
        else:
            raise StopIteration
obj = MyNumbers()
it=iter(obj)
print(next(it))
print(next(it))
print(next(it))
print(next(it))
print(next(it))
    
```

## 4. Python _ Modules

Instead of rewriting them every time, we keep them in a separate file called a module.
A module is a Python file (.py) that contains functions, variables, or classes.
python has buildin modules
math
random
platform
datetime

```python
import modules as mx
print(" the file name is module.py in which we have dict that has persons information ")
person1 = {
    "name": "John",
    "age": 36,
    "country": "Norway"}
print(mx.person1)
print("")
print("Age of person and country")
print(mx.person1["age"])
print(mx.person1["country"])
## Alias(as)
## when we have module nae longer then we can use it

print("")
 ## built in modules 
print("built in modules")
print("print platforms")
import platform
print(platform.system())

## dir() Function
## it tells what module contains 
print("")
print("dir() function in modules")
import platform
print(dir(platform))


## just import function not complete module
print("")
print("just import variable not complete module")
from modules import person1
print(person1["age"])

## import imports the entire module.
## from  import imports only the required function, class, or variable.e
```

```python
print("Current Date or Time")
import datetime
x = datetime.datetime.now()
print(x)

print("")
## print year only
print("print year only")
import datetime
x = datetime.datetime.now()
print(x.year)

print("")
## Week Days Print
print("Print Week Days")
import datetime
x = datetime.datetime.now()
print(x.strftime("%A")) ## strftime() is used to format the date.
## %A represents the full name of the weekday.

## Customize Date
print("")
print("Customize Date")
import datetime
x = datetime.datetime(2020,5,17)
print(x)

## Customize Date and time both
print("")
print("Customize Date and time both")
import datetime
x = datetime.datetime(2025,7,6,10,30,15)
print(x)

## Strftime()
### this function formate the date
print("")
print("Formate the date")
import datetime
x = datetime.datetime.now()
print(x.strftime("%B"))
print("Year :", x.strftime("%Y"))
print("Month :", x.strftime("%B"))
print("Day :", x.strftime("%d"))
print("Week :", x.strftime("%A"))
print("Hour :", x.strftime("%H"))
print("Minute :", x.strftime("%M"))
print("Second :", x.strftime("%S"))
print("AM/PM :", x.strftime("%p"))
```
