## List Comprehension
A shorter syntax for creating a new list based on the values of an existing list.

Creates a new list from an existing list by applying conditions.

```python
fruits=["apple","banana","cherry"]
newlist=[]
for x in fruits:
    if "a" in x:
        newlist.append(x)
print("List Comprehension")
print(newlist)
## Syntax
##[Expression  +  For Loop  +  Optional If Condition]

## another short way
fruit=["apple","banana","cherry","mango"]
newlist=[x for x in fruit if "a" in x]
print("Short Syntax List Comprehension")
print(newlist)

## remove from list 
fruits = ["apple", "banana", "cherry"]
newlist = [x for x in fruits if x != "apple"]
print("Remove from list")
print(newlist)

## if condition 
numbers = [1,2,3,4,5,6]
newlist = [x for x in numbers if x > 3]
print("If condition list comprehension")
print(newlist)

## change  the expression 
fruits = ["apple", "banana", "mango"]
newlist = [x.upper() for x in fruits]
print("Uppercase list comprehension")
print(newlist)

fruits = ["apple", "banana", "mango"]
newlist = ["Hello" for x in fruits]
print(newlist)

## ifelse in expression 
fruits = ["apple","banana","cherry"]
newlist = [x if x != "banana" else "orange" for x in fruits] ## apple not equal to banana condition false

print(newlist)
```

## Python Sort List 
sort the list alphanumerically, ascending by default

```python
fruits = ["orange", "mango", "kiwi", "pineapple", "banana"]
fruits.sort()
print("Sorted list")
print(" ")
print(fruits)

## Descending order 
numbers = [100, 50, 65, 82, 23]
numbers.sort(reverse=True)
print("Sorted list in Descending")
print(" ")
print(numbers)

## Customized sort 
## it  will sort according to our own choice
## abs means absolute value negative to positive 
def myfunc(n):
    return abs(n - 50)
numbers = [100, 50, 65, 82, 23]
numbers.sort(key=myfunc)
print("Customized Sorted list")
print(numbers) ## it will check the difference of each number from 50 then sort
## smallest distance value first then next so on 

## Case Sensitive Sorting 
fruits = ["banana", "Orange", "Kiwi", "cherry"]
fruits.sort()
print("Case Sensitive Sorted list")
print(fruits)  ## first capital letters then small letters

## Case Insensitive Sorting
fruits = ["banana", "Orange", "Kiwi", "cherry"]
fruits.sort(key=str.lower) ## low and uppercase treated same 
print(fruits)
```

## Python _ Join Set
1. Union()
2. Intersection()
3. Difference
4. Symmetric difference

```python
### Union
set1 = {"apple", "banana"}
set2 = {"orange", "kiwi"}
set3 = set1.union(set2)
print("Union of two sets")
print(set3)

print(" ")

## Using |  ( this is shortcut method of union)
set1 = {"a", "b"}
set2 = {1,2,3}
set3 = set1 | set2
print("Union of two sets using | ")
print(set3)

print(" ")

## Multiple Set Join
set1 = {"a"}
set2 = {"b"}
set3 = {"c"}
set4 = {"d"}
result = set1.union(set2, set3, set4)
print("Union of multiple sets")
print(result)

print("")

## Join Set with Tuple
x = {"apple", "banana"}
y = ("orange", "kiwi")
z = x.union(y)
print("Union Set with Tuple")
print(z)

## Update method change set permanently original set changes while 
## union doesnot change original set they remain same
print(" ")

## Intersection  ( return commn values)
set1 = {"apple","banana","cherry"}
set2 = {"banana","orange","apple"}
print("Intersection of twoo sets ")
print(set1.intersection(set2))

print(" ")

## Difference (return values that are in a but not in b)
set1 = {"apple","banana","cherry"}
set2 = {"banana","orange"}
print("The difference of two sets")
print(set1.difference(set2))

print(" ")

## Symmetric_difference  (Remove common values)
set1 = {"apple","banana","cherry"}
set2 = {"banana","orange"}
print("Symmetric_differenceof two sets")
print(set1.symmetric_difference(set2))
```

## Python Frozen Set 
it is immutable(Unchangeable) version of set 

```python
x = frozenset({"apple", "banana", "cherry"})
print("Frozen set")
print(x) ## frozenset can not  add remove items

print(" ")
print(type(fruits))

print(" ") 
y=x.copy()
print("Copy of frozen set:")
print(y)

print(" ")
## union of frozen set
a = frozenset({"apple", "banana"})
b = frozenset({"orange", "kiwi"})
print("Union of two frozen sets ")
print(a.union(b))

print(" ")
## intersection of two frozen set
a = frozenset({"apple", "banana"})
b = frozenset({"banana", "orange"})
print("Intersection of two frozen sets")
print(a.intersection(b))

print(" ")
## Difference of two frozenset
a = frozenset({"apple", "banana"})
b = frozenset({"banana", "orange"})
print("Difference of two frozensets")
print(a.difference(b))

print(" ")
## Symmetric_Difference of two frozen set
a = frozenset({"apple", "banana"})
b = frozenset({"banana", "orange"})
print("Symmetric_Difference of two frozen sets")
print(a.symmetric_difference(b))

print(" ")
## issubset() check either 1st set is part of second set
## (a<=b)
a = frozenset({1,2})
b = frozenset({1,2,3,4})
print("Frozen set issubet():")
print(a.issubset(b))

print(" ")
## issuperset() check either all elements of set 2 are present in set 1
## (a>=b)
a = frozenset({1,2,3,4})
b = frozenset({1,2})
print("Frozen set issuperset():")
print(a.issuperset(b))

print(" ")
## isdisjoint() check common elements exists or not
a = frozenset({1,2})
b = frozenset({3,4})
print("Frozen set isdisjoint():")
print(a.isdisjoint(b))
```

```python
#Python Scope

# Scope means the area of a program where a variable is accessible
# It defines where a variable can and cannot be used
x = 20
print(x)

# Here, x is accessible

# But if a variable is created inside a function,
# it cannot be accessed outside the function

# This concept is called scope

#1. Local Scope

# A local variable is available only inside its function
def myfunc1():
    x = 300 #local variable
    print(x)

myfunc1()

# If we try to use a local variable outside the function
def myfunc():
    h = 32
myfunc()

# print(h)  # NameError -> name 'h' is not defined

# Function Inside a Function
def outer34():
    j = 40
    
    def inner():
        print(j)
    inner()
outer34()

# The inner function can access the outer function's variable
# because it is defined inside the outer function

# The inner function itself is not accessible outside
def outer1():
    x=300
    def inner1():
        print(x)
# inner1()  # NameError

#Global Scope

# Now the variable is created outside the function
r = 34

def myfunc3():
    print(r)
myfunc3()

print(r)

# A global variable can be accessed both inside and outside functions

# Local -> Only inside the function
# Global -> Accessible everywhere

# Same Variable Name
d = 90
def my_func():
    d = 34 
    print(d)  # Python first searches inside the function.
    # If found, it stops searching.
    # Otherwise, it searches outside the function.
my_func()
print(d) # Prints the global variable

# Memory -> Global memory and function memory

###global Keyword
def func1():
    global x
    x = 304 # This variable becomes global
func1()
print(x)

# Modify an Existing Global Variable
v=323

def func2():
    global v
    v=230
    
func2()
print(v) #Updated from 323 to 230

#nonlocal Keyword

# Used inside a nested function
# Allows access to and modification of the outer function's variable
def outer2():
    s = "june"
    
    def inner2():
        nonlocal s #this update outer variable
        s = "July" #non-local -> possible use in outer function
    inner2()
    print(s)
outer2()

#without non-local
def outer3():
    p = "june"
    
    def inner3():
        p = "July" 
    inner3()
    print(p)  # Prints the outer variable
outer3()

#LEGB Rule 
# Python's variable lookup order

# Local -> Enclosing -> Global -> Built-in


#L = Local 
#Current Function 
x=5

def test12():
    x=10
    print(x)

test12()

#E = Enclosing 
#Outer Function

def outer12():
    x=20
    def inner():
        print(x)
    inner()

outer12() 
# If not found in Local, Python searches in Enclosing

#G = Global
x=50

def test():
    print(x)

test()

#B = Built-in
# Python's built-in names
numbers=[1,2,3]
print(len(numbers))

# len() is not created by you.
# It is already provided by Python.
# It belongs to the Built-in namespace.


# LEGB Search Order

# Local -> Enclosing -> Global -> Built-in
x = "Global"

def outer():
    x = "Enclosing"
    def inner():
        x = "Local"
        print(x)
    inner()
    print(x)
outer()

print(x)
```

Python Decorator

```python
#Python Decorators

# A decorator adds extra functionality to a function
# without changing the original function's code

#Without Decorator
def greet_new():
    return "Zarmeen"
print(greet_new())
# Now we want the output in uppercase

# Simple Method
def greet_new1():
    return "Hello"

print(greet_new1().upper())

# But if there are 100 functions,
# should we write .upper() for every function?

# To solve this problem, use a decorator

#Basic Decorator

# def function_name(func):

def changecase(func):
    def myinner():
        return func().upper()
    return myinner

# changecase(greet)
# func = greet

# Inner function -> A function defined inside another function

# return func().upper() -> First execute the original function,
# then apply upper()

# return myinner -> Return the function itself

# return myinner() -> Execute the function immediately

###Decorator Usage

@changecase
def greet():
    return "hello123"

greet()

#Python Internally 
def greet1():
    return "hello_new"

greet22 = changecase(greet1)
print(greet22())

###Example
def changecase1(func1):

    def myinner1():
        return func1().lower()

    return myinner1

@changecase1
def greet23():
    return "ZARMEEN ALISHA "

print(greet23())

#Multiple Functions
@changecase
def greeter():
    return "Hello"
greeter()

@changecase
def city():
    return "Lahore"
city()

###Function with Arguments

def change1(func_new):
    def inner_new(name):
        return func_new(name).upper()
    return inner_new

@change1
def new1(name):
    return "Hello " + name

print(new1("zarmeen"))

# Problem -> If a function has two or more arguments,
# the wrapper must be changed every time

# Solution -> *args and **kwargs

### *args and **kwargs

# Universal wrapper for decorators

# def innerfunc1(*args, **kwargs):
#     return new_func1(*args, **kwargs).upper()

###Example
def changing(newfunc1):
    def inner11(*email, **data):
        return newfunc1(*email, **data).upper()
    return inner11

@changing
def data(email, data):
    return "hi " + email + data

print(data("@gmail.com ", data = "alisha" ))

#Decorators with Arguments

# Suppose you want to choose uppercase or lowercase

# @changecase(1)
# or
# @changecase(2)

# For this, create one more wrapper

#Decorator Argument
#↓
#Decorator
#↓
#Wrapper
#↓
#Original Function

##Example
def alisha1(choice):
    def decorator(func):
        def wrapper():
            if choice == 1:
                return func().upper()
            
            else:
                return func().lower()
        return wrapper
    return decorator

#use
@alisha1(1)
def string11():
    return "zarmeen_alisha "

print(string11())

@alisha1(2)
def string2():
    return "hello1233we"

print(string2())

#Multiple Decorators

# If multiple decorators are applied,
# the decorator closest to the function runs first

def addgreeting(func):
    def wrapper():
        return "Hello " + func()
    return wrapper

def changecase(func):
    def wrapper():
        return func().upper()
    return wrapper

@changecase
@addgreeting
def greet():
    return "Zar"

print(greet())

# Rule: Decorators execute from bottom to top.
# The decorator closest to the function executes first.

###Function Metadata

# Every function has metadata

def new1():
    pass

print(new1.__name__)

# After applying a decorator
@changecase
def new2():
    pass
print(new2.__name__)

# Why does it print wrapper instead of new2?
# Because the decorator replaces the original function
# with the wrapper function.

# functools.wraps
import functools
def case1(func_new):
    @functools.wraps(func_new)
    
    def inner_new():
        return func_new().upper()
    
    return inner_new

@case1
def greett():
    return "abc"

print(greett.__name__) # The original function name is preserved.

#Original Function
#   ↓
#Decorator Receive
#   ↓
#Wrapper Function
#   ↓ 
#Extra Work
#   ↓
#Original Function
#   ↓
#Modified Result
#   ↓
#Return

##Examples
#Example 1: Login Check

def login_required(login_func):
    def inner_func(gmail):
        return login_func(gmail).lower()
    return inner_func

@login_required
def profile(gmail):
    return gmail + "@gmail.com"

print(profile("Zarmeen12345"))


#Example 2: Timing
#@timer
#def train_model():

#Example 3: Permission
#@admin_only
#def delete_user():
```

Lambda Functions

```python
#Lambda Functions

# Anonymous (nameless) short functions

# Normal Function
def add(a,b):
    return a+b

print(add(3,4))

#lambda function
add = lambda a,b : a + b
print(add(3,42))

# lambda a, b: a + b is equivalent to:
# def add(a, b):
#     return a + b

# Syntax -> lambda arguments : expression

# General Form -> lambda x: x * 2

# Example 
square = lambda a : a*a
print(square(5))

cube = lambda b : b*b*b
print(cube(3))

#Multiple Arguments
multiply = lambda a,b : a*b
print(multiply(5,6))

#Three Arguments
sum = lambda a,b,c : a + b + c
print(sum(89,34,21))

# When should we use a lambda function?

# Use lambda when:
# - The function is temporary
# - It contains only one expression
# - It is used only once

# If the function is long or complex, use def instead


#Lambda Returning a Function
def a_func(n):
    return lambda a: a*n

value1 = a_func(9) # n = 9 -> value1 = lambda a: a * 9
print(value1(8)) 

###Lambda with map()

# Suppose we want to double all the numbers

numberss = [1,2,3,4]

#without lambda
result = []
for i in numberss:
    result.append(i*2)
print(result)

#lambda + map

result = list(map(lambda x: x*2,numberss))
print(result)

# map() applies the function to every element

# Lambda with filter()

# We want only even numbers
even = list(filter(lambda x : x%2==0, numberss ))
print(even)
# filter() returns only the elements that satisfy the condition

#lambda with sorted()
students=[("Ali",20),("Ahmed",18),("Sara",22)]

sorted_students = sorted(students,key=lambda x:x[1])
print(sorted_students)

# sorted() uses the lambda function to decide the sorting key
```

Python Recursion

```python
#Recursion

# Recursion means a function calls itself

def hello():
    print("hello")
    hello()  # The function calls itself -> infinite recursion

#hello()

# Base Case

# Every recursive function must have a stopping condition.
# This stopping condition is called the base case.

def hello1(n):
    if n == 0:
        print("Done")
    else:
        print(n)
        hello1(n-1)
hello1(8)

# Factorial -> 5! = 5 × 4 × 3 × 2 × 1

# Recursive solution
def factorial(n):
    if n == 1:
        return 1
    return n*factorial(n-1)

print(factorial(5))

# Fibonacci

# Rule -> The next number is the sum of the previous two numbers

# 0, 1, 1, 2, 3, 5, 8, 13, 21

def fibonacci(n):
    if n<=1:
        return n
    return fibonacci(n-1)+fibonacci(n-2)

print(fibonacci(5))

#fib(3) = fib(2) + fib(1)
#       = 1 + 1

#Fibonacci sequence
# n      Result
# 0  →   0
# 1  →   1
# 2  →   1
# 3  →   2
# 4  →   3
# 5  →   5
# 6  →   8
# 7  →   13

##Recursion Limit

# By default, Python allows approximately 1000 recursive calls
import sys
print(sys.getrecursionlimit()) 
#set limit

sys.setrecursionlimit(4000)
print(sys.getrecursionlimit())
```

```python
#Normal Function 
def number11():
    return [1,2,3]

# The entire list is created and stored in memory

#Generator
def num1():
    yield 1
    yield 2
    yield 3

# No list is created.

# Values are generated one by one.

# This makes generators memory-efficient.

##yield vs return

# return -> Ends the function immediately
def test1():
    return 5  # Function ends here
    return 10  # This line is never executed

print(test1()) #output -> 5

# yield -> Pauses the function instead of ending it
def test2():
    yield 1
    yield 2
    yield 3

for i in test2():
    print(i)

#next()

def test3():
    yield "zarmeen"
    yield "alisha"
    yield "Ms.samra"

x = test3()
print(next(x))
print(next(x))
print(next(x))
# print(next(x))  # ERROR -> StopIteration (No more values are available)

###Memory Saving

# Suppose we need 10 million numbers

# List

# Expression -> [x * x for x in range(5)]
number123 = [x for x in range(10000000)] # The entire list is stored in RAM 

# Generator

# Expression -> (x * x for x in range(5))
# It does not create a list; it returns a generator object
number456 = (x for x in range(10000000))
# Only one value is generated at a time.

# This is why generators are very memory-efficient.

# Convert to a list
list(number456)


#yield
# ↓
#Pause
# ↓
#Resume
# ↓
#Memory save
# ↓
#Returns an iterator

#Lambda vs Normal Function

# Normal Function           Lambda
# def is used               lambda is used
# Multiple statements       Only one expression
# Has a name                Anonymous
# Suitable for large code   Suitable for small functions

#Generator vs Normal Function

# Normal Function                 Generator
# return                          yield
# Returns all values at once      Returns values one by one
# Function ends                   Function pauses
# Uses more memory                Uses less memory

#Recursion vs Loop

# Loop                            Recursion
# Uses for/while                  A function calls itself
# Faster                          Can be slower in some cases
# Easier                          Slightly more difficult
# Does not use the call stack     Uses the function call stack
```
