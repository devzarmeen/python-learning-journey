They define the behavior of objects created from the class.

**Python Inheritance**

Inheritance allows us to define a class that inherits all the methods and properties from another class.

**Parent Class**

it is class being inherited from, also called base class.

class Person:

**Child Class**

It is the class that inherits from another class, also called derived class.

class Student(Person):

**Parent Class**

```python
## Parent class
class Person:
    def __init__(self,fname,lname,):
        self.firstname=fname
        self.lastname=lname


    def printname(self):
        print(self.firstname,self.lastname)
## Child class
class Student(Person):
    pass   ## pass used child does not has its own properties

x=Student("Zarmeen","Alisha")
x.printname()

```

```python
## if child has its own _int_() method then parents _int_() method will stop and child's method will work
print("")
print("Add Property")
## Parent Class
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname
    def printname(self):
        print(self.firstname, self.lastname)


## Child Class
class Student(Person):
    def __init__(self, fname, lname):
        super().__init__(fname, lname)   # Call Parent Constructor
        self.graduationyear = 2019        # Child's own property

## Create Object
x = Student("Zarmeen", "Alisha")
## Print Parent Data
x.printname()
## Print Child Property
print("Graduation Year:", x.graduationyear)
```

```python
## Method Add
print("Add method")
## Parent Class
class Person:
    def __init__(self, fname, lname):
        self.firstname = fname
        self.lastname = lname

    def printname(self):
        print(self.firstname, self.lastname)

## Child Class
class Student(Person):

    def __init__(self, fname, lname, year):
        super().__init__(fname, lname)
        self.graduationyear = year

    def welcome(self):
        print("Welcome", self.firstname, self.lastname)
        print("Graduation Year:", self.graduationyear)

## Create Object
x = Student("Zarmeen", "Alisha", 2026)
## Parent Method
x.printname()
print("")
## Child Method
x.welcome()
```

```python
## Method Override
print("Method Override")
print("")
## Parent Class
class Person:
    def display(self):
        print("I am a Person")

## Child Class
class Student(Person):
    def display(self):
        print("I am a Student")
## Create Object
s1 = Student()
## Call Method
s1.display()
```

**4.Python Polymorphism**

Polymorphism means many forms and in programming it refers to methods/functions/operators with same name that can be executed on many objects or class  

**4.1Function Polymorphism**  

Some built-in functions in Python are already polymorphic.

```python
## String counts characters
print("String Example")
print(len("Hello"))

## Tuple count items
print("")
print("Tuple example")
mytuple=("apple","banana","cherry")
print(len(mytuple))

##Dictionary count keys
print("")
print("Dictionary Example")
thisdict={
"brand":"Ford",
"model":"Mustang",
"year":1964
}
print(len(thisdict))

## len() name same behavior different for each
```

**4.2Class Polymorphism**

```python
## Car
class Car:
    def move(self):
        print("Car")
        print("Drive!")
## Boat
class Boat:
    def move(self):
        print("Boat")
        print("Sail!")
## Plane
class Plane:
    def move(self):
        print("Plane")
        print("Fly!")
## move() method are same but output are different

## objects
car1=Car()
boat1=Boat()
plane1=Plane()

## For Loop

for x in (car1,boat1,plane1):
    print("")
    x.move()
```

**4.3Inheritance**

```python
class Vehicle:
    def __init__(self,brand,model):
        self.brand=brand
        self.model=model
    def move(self):
        print("Move!")
## Car
class Car(Vehicle):
    pass   ## vehicle ki properties inherit kr rha ha
## Boat
class Boat(Vehicle):
    def move(self):
        print("Sail!")
## Plane
class Plane(Vehicle):
    def move(self):
        print("Fly!")

## objects
car1=Car("Ford","Mustang")
boat1=Boat("Ibiza","Touring 20")
plane1=Plane("Boeing","747")

## Loop
for x in (car1,boat1,plane1):
    print("")
    print(x.brand)
    print(x.model)
    x.move()
```

**5.Python Encapsulation**

Protecting data inside a class

It means keeping data (properties/varaiables) and methods together in a class, while controlling how the data can be accessed from outside the class.

This prevents accidental changes to your data and hides the internal details of how your class works.

**5.1 Private Property**

In Python, we can make properties private by using a double underscore __ prefix:

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.__age = age
p1 = Person("Alisha",20)
print(p1.name)
##print(p1.__age)   it will giver attribute error
print(" ")

## Getter Method
print("Access private variable using get method")
class Person:
    def __init__(self,name,age):
        self.name=name
        self.__age=age
    def get_age(self):
        return self.__age
p1=Person("Alisha",20)
print(p1.get_age())

print("")
## Setter Method
## if we want to change .__age it is not allowed directly
print("Set age")
class Person:
    def __init__(self,name,age):
        self.name=name
        self.__age=age
    def get_age(self):
        return self.__age
    def set_age(self,age):
        if age>0:
            self.__age=age
        else:
            print("Age must be positive")
p1=Person("Ali",20)
print("Age before set")
print(p1.get_age())
p1.set_age(25)
print("Age after set")
print(p1.get_age())

## single underscore(_) is protected it can be accessed directly while double underscore(__)
  # is protected it can not be accessed directly to access this we need get method
```

**5.2 Private Method**

```python
print("Private Method")
class Calculator:     ## class
    def __init__(self):    ## constructor
        self.result=0      ## property
## private Method
    def __validate(self,num):
        if isinstance(num,(int,float)):    ## isinstance check type of values
            return True
        return False
## add method
    def add(self,num):
        if self.__validate(num):
            self.result+=num
        else:
            print("Invalid Number")
## object create
calc=Calculator()
## call add()
calc.add(10)
## print results
print(calc.result)
```

**6.Python Inner Class**

An inner class is a class defined inside another class. The inner class can access the properties and methods of the outer class.

Inner classes are useful for grouping classes that are only used in one place, making your code more organized.

**Synatx**

class Outer:

    class Inner:
        pass

```python
class Outer:
    def __init__(self):
        self.name="Outer Class"
    class Inner:
        def __init__(self):
            self.name="Inner Class"
        def display(self):
            print("This is the inner class")
outer=Outer()
print(outer.name)

## inner access properties of outer
print("")
print("Inner access properties of outer")
class Outer:
    def __init__(self):
        self.name="Outer"

    class Inner:
        def __init__(self):
            self.name="Inner"
        def display(self):
            print("Hello from inner class")

outer=Outer()
inner=outer.Inner()
inner.display()
```

**Multiple Inner Classes**

```python
class Computer:
  def __init__(self):
    self.cpu = self.CPU()
    self.ram = self.RAM()

  class CPU:
    def process(self):
      print("Processing data...")

  class RAM:
    def store(self):
      print("Storing data...")

computer = Computer()
computer.cpu.process()
computer.ram.store()
```
