## 3. Matplotlib Pyplot

## Pyplot

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

## What is a Submodule?

Matplotlib is a library.

A library contains smaller components.

These components are called submodules.

Matplotlib
      │
      ├── pyplot
      ├── animation
      ├── colors
      ├── patches

Pyplot package can be referred to as plt

```python
## line in a diagram from position (0,0) to position (6,250):

import matplotlib.pyplot as plt   ## to draw a graph
import numpy as np  ## to form an array of numbers

xpoints = np.array([0, 6])       ## value on x-axis 
ypoints = np.array([0, 250])     ## value on y-axis

plt.plot(xpoints, ypoints)      ## Connects the X and Y points with a line to create a graph.
plt.show()                      ## display graph on screen
```

## 4. Plotting Without Line

## Plotting x and y points

# plot()

The plot() function is used to draw points (markers) in a diagram.

By default, the plot() function draws a line from point to point.

The function takes parameters for specifying points in the diagram.

Parameter 1 is an array containing the points on the x-axis.

Parameter 2 is an array containing the points on the y-axis.

If we need to plot a line from (1, 3) to (8, 10), we have to pass two arrays [1, 8] and [3, 10] to the plot function.

```python
## Draw a line in a diagram from position (1, 3) to position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

The x-axis is the horizontal axis.

The y-axis is the vertical axis.

## Plotting Without Line

To plot only the markers, we can use shortcut string notation parameter 'o', which means 'rings'

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8])
ypoints = np.array([3, 10])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

## Mltiple Points

we can plot as many points as you like

just make sure we have the same number of points in both axis.

```python
## Draw a line in a diagram from position (1, 3) to (2, 8) then to (6, 1) and finally to position (8, 10):

import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 2, 6, 8])
ypoints = np.array([3, 8, 1, 10])

plt.plot(xpoints, ypoints)
plt.show()
```

## Deafult X-Points

If we do not specify the points on the x-axis, they will get the default values 0, 1, 2, 3 etc., depending on the length of the y-points.

```python
## Plotting without x-points:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10, 5, 7])

plt.plot(ypoints)
plt.show()
```

The Default x-points in the  above example are [0, 1, 2, 3, 4, 5].

## 5. Matplotlib Markers

## Markers 

We can use the keyword argument marker to emphasize each point with a specified marker.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3,8,1,10])

plt.plot(ypoints, marker='o')

plt.show()
```

## Format String fmt

we can also use the shortcut string notation parameter to specify the marker.

This parameter is also called fmt, and is written with this syntax:

marker|line|color

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

## Line Reference

    '-.'	            Dashed/dotted line

## Marker Size 

keyword argument markersize or the shorter version, ms to set the size of the markers:

```python
## Set size of marker to 20
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints,marker = '<', ms = 20, color='y')
plt.show()
```

## Marker Color

```python
## Set the EDGE color to red
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'h', ms = 20, mec = 'm')
plt.show()
```

## Marker Face Color 

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 's', ms = 20, mfc = 'r')
plt.show()
```

## Marker with facecolor

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = '2', ms = 20, mfc = 'g', mec='g')
plt.show()
```

## 6. Matplotlib Line

## Linestyle

Keyword argument linestyle, or shorter ls, to change the style of the plotted line:

```python
## Use a dotted line:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted', color="r")
plt.show()
```

## Dashed line

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed', color="g")
plt.show()
```

## Shorter Syntax

linestyle can be written as ls.

dotted can be written as :.

dashed can be written as --.

```python
import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':', color="r")
plt.show()
```

## Line Width

The keyword argument linewidth or the shorter lw to change the width of the line.

The value is a floating number, in points:

```python
## Plot with a 20.5pt wide line:

import matplotlib.pyplot as plt
import numpy as np

ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

## Multiple Lines

```python
## Draw two lines by specifying a plt.plot() function for each line:

import matplotlib.pyplot as plt
import numpy as np

y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1)
plt.plot(y2)

plt.show()
```

##  Draw two lines by specifiyng the x- and y-point values for both lines

```python
import matplotlib.pyplot as plt
import numpy as np

x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

plt.plot(x1, y1, x2, y2)
plt.show()
```
