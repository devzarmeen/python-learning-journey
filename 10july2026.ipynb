**1.Python File Handling**

It is an important part of any web application

Python has several functions for creating,reading, updating and deleting files

Key function for working with files in python is open() function

open() function takes two parameters:

1.filename

2.Mode

There are four different method(modes) for  opening file:

"r" - Read - Default
value.

"a" - Append

"w" - Write

"x" - Create


**f = open("student.txt")**
      opened file will be stored in f, and now we can work on file through f

**1. Read Mode("r")**

      f = open("demo.txt", "r")

      it will read file if exists if does not exists it will give error "File Not Found Error"
**2. Write Mode("w")**
    
      f=open("demo.txt","w")
      f.write("Python")

      write in file if file already exists it will remove the previous data and will write new data in file

**3. Append Method("a")**

      f=open("demo.txt","a")
      f.write("Alisha")

      it will add new data in existing file and will not change or remove already stored data

**4. Create Method("x")**

      f = open("demo.txt","x")

      Create the file if it does not exists

      But if file already exists it will give error "FileExistsError"

**Text Mode("t")**

      open("demo.txt","rt")

      t means text file

**Binary Mode("b")**

      open("demo.jpg","rb")

      rb means read binary

By default mode is text and reading mode

2. Python File Open 

```python
f = open("10julyStudent.txt")    ## Main file and txt file both should be in same folder
print(f.read())  ## read() function is used to read the content of the file 

## if file is in another folder then we have to give the path of the file in open() function. 

# f = open(r"E:6thSemester\\Student.txt")    \\ slash indicates escape character and r indicates raw string.
 # print(f.read())
 
## with Statement is used to open a file and automatically close it after the block of code is executed.
print("")
print("Using with statement")
with open("10julyStudent.txt") as f:
    print(f.read())
    
## close() function is used to close the file manually after reading the content of the file.
print("")
print("Using close() function")
f=open("10julyStudent.txt")
print(f.read())
f.close()

## It is necessary to close file after reading the content of the file 
# otherwise it will occupied memory, Waste resources and create problem in future.

## read(5) function is used to read the first 5 characters of the file.
print("")
print("Using read(5) function")
with open("10julyStudent.txt") as f:
    print(f.read(5))
    
## readline() function is used to read the first line of the file.
print("")
print("Using readline() function")
with open("10julyStudent.txt") as f:
    print(f.readline())
    
## Do readline() function is used to read the first 2 lines of the file.
print("")
print("Using readline() function to read first 2 lines")
with open("10julyStudent.txt") as f:
    print(f.readline())
    print(f.readline())
    
## read through loop is used to read the content of the file line by line.
print("")
print("Using for loop to read the content of the file line by line")
with open("10julyStudent.txt") as f:
    for x in f:
        print(x)
```

3. Python File Write

    1. append mode(a)      will append to the end of the file
    
    2. write mode(w)       will overwrite any existing content

3.1 Append Mode ("a")

```python
with open("10julyStudent.txt", "a") as f:
    f.write("Now the file has more content!")
    print("Content added successfully!")
    
print("")
### Now read the content of the file after adding new content.
print("Reading the content of the file after adding new content")
with open("10julyStudent.txt") as f:
    print(f.read())
```

3.2 Write Mode ("w")

```python
with open("10julydemofile.txt","w") as f:
    f.write("Woops! I have deleted the content!")
    print("Content deleted successfully!")
    
print("")
### Now read the content of the file after deleting the content.
print("Reading the content of the file after deleting the content")
with open("10julydemofile.txt") as f:
    print(f.read())
```

3.3Create New File ("x")

```python
print("Creating a new file")
f=open("10julynewfile.txt","x")
print("New file created successfully!")

## if file already exists then it will give an error. 
## a and w are used to create a new file if it does not exist. they will not give an error if file already exists.
```
