**1. Python String Formatting**

Insert values and varaiables in string

**1.1F-String Method**

it is easy and fast method

Syntax = f"Text{varaiable}"

f indicates F-string and {} is placeholder for value of variable

```python
name = "Alisha"
print("String ")
print("Hello", name)
## if we want complete sentence then we use String Formatting

print("")
## F-String (Recommended)
name = "Alisha"
txt = f"My name is {name}"
print("F-String of variable name")
print(txt)

print("")
## F-String for Multiple Variables
name = "Alisha"
age = 22
city = "Lahore"
print("F-String for Multiple Variables")
print(f"My name is {name}. I am {age} years old. I live in {city}.")

print("")
## Decimal Formatting  if  we want 2 decimal places
price = 59
print("Decimal Formatting")
print(f"The price is {price:.2f}")  ## :  indicates formatting has started and .2 indicates 2 decimal places

print("")
## Math Operation
print("perform multiplication using f-string")
print(f"{20*5}")

print("")
## Addition
a = 20
b = 10
print("perform addition using f-string")
print(f"Sum = {a+b}")

print("")
## Tax Example
price = 100
tax = 0.10
print("perform tax using f-string")
print(f"Total = {price + price*tax}")

print("")
## If Else inside F-String
price = 49
print("if else inside f-string")
print(f"It is {'Expensive' if price>50 else 'Cheap'}")

print("")
## Function Inside F-String
fruit = "apple"
print("Function Inside F-String")
print(f"I like {fruit.upper()}")

print("")
## User Function
def square(x):
    return x*x
print("User Function ")
print(f"Answer = {square(5)}")

print("")
## Thousand Separator
price=1500000
print("Thousand Separator")
print(f"price={price:,}")

print("")
## Alignment  left Align
name="Alisha"
print("Alignment left Align")
print(f"Left Align={name:<10}End")

print("")
## Alignment right Align
name="Alish"
print("Alignment right Align")
print(f"right Align={name:>10}End")

print("")
## Alignment Center Align
name="Alisha"
print("Alignment Center Align")
print(f"Center Align=start{name:^10}end")

print("")
## Percentage
marks = 0.85
print("Show percentage of marks")
print(f"{marks:%}")

print("")
## Binary
num=10
print("Convert num into binary")
print(f"{num:b}")

print("")
## octal
num=10
print("Convert num into octal")
print(f"{num:o}")

print("")
## Hexadecimal
num=25
print("Convert num into hexadecimal")
print(f"{num:o}")
```

**7.Python _int_() Method**

All classes have a built-in method called __init__(), which is always executed when the class is being initiated,or object is being created.

The __init__() method is used to assign values to object properties, or to perform operations that are necessary when the object is being created.

```python
class Person:
    def __init__(self):
        print("Object Created")
p1 = Person()    ## we dont call _int_() , we just write person() it automatically calls  _int_()

print("")
print("Example with _int_()")
## Example  with _int_()
class Person:      ## class
    def __init__(self, name, age):   ## constructor  executed on creation of object
        self.name = name   ## self represent current object
        self.age = age
p1 = Person("Zarmeen",36)   ## object created
print(p1.name)
print(p1.age)
## with _int_() Python automatically stores values

print("")
print("Example without _int_()")
## Example  without _int_()
class Person:
    pass
p1 = Person()     ## we have to write manually after each object
p1.name = "Ali"
p1.age = 20
print(p1.name)
print(p1.age)


print("")
print("Multiple Parameters")
## Multiple Parameters
class Person:
    def __init__(self,name,age,city,country):
        self.name=name

        self.age=age

        self.city=city

        self.country=country
p1=Person("Alisha",20,"Lahore","Pakistan")

print(p1.name)

print(p1.age)

print(p1.city)

print(p1.country)
```

**8.Python Self Parameter**

Self parameter is a reference to current object of class

it is used to access properties and  method that belongs to the class

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    def greet(self):
        print("Hello, my name is", self.name, "And My age is",self.age)
p1 = Person("Alisha",20)
p1.greet()

## Without self
print("")
print("WithOut self")
class Person:
    def __init__(name):
        name=name   ## python will be confused which objects name is this

## self name can be  changed and self is always first parameter
print("")
print("Access Multiple properties and change name of self")
## Access Multiple properties
class Car:

    def __init__(self,brand,model,year):

        self.brand=brand
        self.model=model
        self.year=year
    def display(self):
        print(self.brand)
        print(self.model)
        print(self.year)
car1=Car("Toyota","Corolla",2020)
car1.display()

## Calling One Method from Another
print("")
print("Calling One Method from Another")
class Person:
    def __init__(self,name):
        self.name=name
    def greet(self):
        return "Hello " + self.name
    def welcome(self):
        print(self.greet())
p1=Person("Ali")
p1.welcome()
```
