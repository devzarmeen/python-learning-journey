```python
## Highest lowest and average of a list
marks=[78,45,90,67,88]
highest=max(marks)
lowest=min(marks)
average=sum(marks)/len(marks)
print("Highest",highest)
print("Lowest",lowest)
print("Average",average)
```

```python
## remove duplicate values from list
list1=[1,2,2,3,4,4,5]
unique_numbers = list(set(list1))
print("Original list:",list1)
print("List after remoning duplicates:",unique_numbers)
```
