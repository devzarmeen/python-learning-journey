1. Numpy Module
NumPy is a Python library used for working with arrays.

It also has functions for working in domain of linear algebra, fourier transform, and matrices.

NumPy was created in 2005 by Travis Oliphant. It is an open source project and you can use it freely.

NumPy stands for Numerical Python.

NumPy is a Python library and is written partially in Python, but most of the parts that require fast computation are written in C or C++.

Why Use Numpy 

In Python we have lists that serve the purpose of arrays, but they are slow to process.

NumPy aims to provide an array object that is up to 50x faster than traditional Python lists.

The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy.

Arrays are very frequently used in data science, where speed and resources are very important.

Data Science:
  It is a branch of computer science where we study how to store, use and analyze data for deriving information from it.

Why is Numpy Faster Than Lists

NumPy arrays are stored at one continuous place in memory unlike lists, so processes can access and manipulate them very efficiently.

This behavior is called locality of reference in computer science.

This is the main reason why NumPy is faster than lists. Also it is optimized to work with latest CPU architectures.

Attributes & data types

.shape — dimensions as a tuple  (No of rows and columns)

.ndim — number of dimensions

.dtype — element type

.astype() — cast to a new dtype

```python
import numpy as np
a = np.array([[1, 2, 3], [4, 5, 6]])
print("Original Array:")
print(a)
# Shape
print("\nShape of Array:")
print(a.shape) ## (2, 3)

# Number of Dimensions
print("\nNumber of Dimensions:")
print(a.ndim) ## 2

# Data Type
print("\nData Type:")
print(a.dtype)   # dtype('int64' or int32 both are same)

# Convert Data Type
b = a.astype(np.float32)
print("\nArray After Converting to float32:")
print(b)

print("\nNew Data Type:")
print(b.dtype)
```

Numpy Slicing

taking elements from one given index to another given index.

Synatax:
array[start:end]   From which index it is starting  and At which index it will stop but stop value will not beincluded
        or 
array[start:end:step] 

```python
import numpy as np
arr = np.array([1,2,3,4,5,6,7])
print("Slicing Array:")
print(arr[1:5])

print("")
print("Slicing Array with start value only:")   
print(arr[4:])

print("")
print("Slicing Array with end value only:")   
print(arr[:4])

print("")
print("Slicing Array without start and end value:")
print(arr[:])

print("")
print("Slicing Array with negative index:")
print(arr[-3:-1])
```

Numpy Copy vs View

```python
## copy() method creates a new array that is a copy of the original array, and any changes made to the copy will not affect the original array. This is useful when you want to create a new array that is independent of the original array.
import numpy as np
arr=np.array([1,2,3,4]) ## original array should not be affected by changes made to the copy of the array. This is because the copy() method creates a new array that is a copy of the original array, and any changes made to the copy will not affect the original array.
copy_arr=arr.copy()
arr[0]=100
print("Original Array after changing first element:")
print(arr)
print("")
print("Copy of Original Array:")
print(copy_arr) 
## in this sceanrio original array changed but copy of original array is not changed because we have used copy() method to create a copy of original array. If we use view() method then both original and copy of original array will be changed because view() method creates a view of original array and any changes made to the view will affect the original array.

## view() method creates a view of the original array, which means that any changes made to the view will affect the original array. This is useful when you want to create a new array that is dependent on the original array.
import numpy as np
arr=np.array([1,2,3,4])
view_arr=arr.view()
arr[0]=100
print("Original Array after changing first element:")
print(arr)
print("")
print("View of Original Array:")
print(view_arr)
print("")
## base attribute returns the base object of the array. If the array owns its own data, then the base attribute will be None. If the array is a view of another array, then the base attribute will be a reference to the original array.
import numpy as np
arr=np.array([1,2,3])
x=arr.copy()
y=arr.view()
print(x.base)
print(" ")
print(y.base)
```

Numpy Reshape it changes shape but data will remain same 

```python
import numpy as np
arr=np.array([1,2,3,4,5,6])
print("Original 1D Array:")
print(arr)
print("")
print("Reshaping 1D Array to 2D Array:")
newarr=arr.reshape(2,3)
print(newarr)   ## total elements should be same in original and reshaped array. Here, original array has 6 elements and reshaped array also has 6 elements (2*3=6). If we try to reshape the array into a shape that does not have the same number of elements as the original array, we will get an error.

print("")
## Flattening array means converting a multi-dimensional array into a 1D array. We can use flatten() method to flatten the array. It returns a copy of the original array collapsed into one dimension.
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print("Original 2D Array:")
print(arr)
flat=arr.reshape(-1)
print("Array after Flattening:")
print(flat)
```

Numpy Iterating

```python
## 1D Array
import numpy as np
arr=np.array([10,20,30])
print("iterating through 1D array: ")
for x in arr:
    print(x)
    
print("")
## 2D Array
print("iterating through 2D array: ")
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
for x in arr:
    print(x)
    
print("")
## nditer() Array   it is shortcut for n-dimensional iterator. It is used to iterate through an n-dimensional array. It is more efficient than using nested for loops to iterate through an n-dimensional array.
print("iterating through nditer array: ")
import numpy as np
arr=np.array([[1,2],[3,4]])
for x in np.nditer(arr):
    print(x)
    
print("")
## ndenumerate() Array   it is shortcut for n-dimensional enumerator. It is used to iterate through an n-dimensional array and return the index and value of each element in the array. It is more efficient than using nested for loops to iterate through an n-dimensional array.
print("iterating through ndenumerate array: ")
import numpy as np
arr=np.array([[10,20],[30,40]])
for index,value in np.ndenumerate(arr):
    print(index,value)
```

Joining Array Combine 2 or more arrays

```python
## concatenate() method is used to join two or more arrays along an existing axis. It takes a sequence of arrays as input and returns a new array that is the concatenation of the input arrays. The axis parameter specifies the axis along which the arrays will be joined. If axis=0, the arrays will be joined vertically (row-wise), and if axis=1, they will be joined horizontally (column-wise).
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
arr = np.concatenate((arr1,arr2))
print("Concatenated Array:")
print(arr)

print("")
## 2D Array Concatenation
import numpy as np
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6]])
arr = np.concatenate((arr1,arr2),axis=0)
print("Concatenated 2D Array:")
print(arr)

print("")
## Stack() method is used to join two or more arrays along a new axis. It takes a sequence of arrays as input and returns a new array that is the concatenation of the input arrays along a new axis. The axis parameter specifies the axis along which the arrays will be joined. If axis=0, the arrays will be joined vertically (row-wise), and if axis=1, they will be joined horizontally (column-wise).
import numpy as np
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.stack((arr1,arr2))
print("Stacked Array:")
print(arr)

print("")
## hstack() method is used to join two or more arrays horizontally (column-wise). It takes a sequence of arrays as input and returns a new array that is the concatenation of the input arrays along the second axis (axis=1).
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
print("Horizontally Stacked Array:")
print(np.hstack((a,b)))

print("")
## vstack() method is used to join two or more arrays vertically (row-wise). It takes a sequence of arrays as input and returns a new array that is the concatenation of the input arrays along the first axis (axis=0).
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
print("Vertically Stacked Array")
print(np.vstack((a,b)))

print("")
## dstack() depth stack it creates third dimension
import numpy as np
a=np.array([1,2,3])
b=np.array([4,5,6])
print("Depth STacked Array")
print(np.dstack((a,b)))
```

Numpy Filter Array

```python
## Boolean Indexing
import numpy as np
arr=np.array([41,42,43,44])
x=[True,False,True,False]
newarr=arr[x]
print("Boolean Filtering")
print(newarr)

print("")
print("Manual Filtering")
## Manual Filtering
import numpy as np
arr=np.array([41,42,43,44])
filter_arr=[]
for element in arr:
    if element>42:
        filter_arr.append(True)
    else:
        filter_arr.append(False)
print(filter_arr)
newarr=arr[filter_arr]
print("New Array")
print(newarr)

print("")
print("Even Number Filtering")
## Even Number
import numpy as np
arr=np.array([1,2,3,4,5,6,7])
filter_arr=arr%2==0
print(filter_arr)
newarr=arr[filter_arr]
print("New Array")
print(newarr)
```
