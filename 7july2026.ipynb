**Python _ Math**

To perform calculations on number there are two method
1. built-in math functions (no need to import these)
2. Math Module( need to import like import math)

**1. Built in Math functions:**


```python
## 1. min() Function
## this function will return minimum number
## syntax = min(num1,num2,num3)

x = min(5, 10, 25)
print("Minimum value is:")
print(x)

print(" ")
## 2. max()Function
## this function will return maximum number/value
## syntax = max(num1,num2,num3)
y = max(5, 10, 25)
print("Maximum number is:")
print(y)

## 3. abs() Function
## it convert negative number into positive number
## syntax = abs(number)
print(" ")
print("Absolute value is ")
x = abs(-7.25)
print(x)
## or
x=-10
print(abs(-10))  ## if we have positive number then there will be no change


## 4. pow() Function
## it calculate power
## syntax= pow(base,power)
print("")
print("Power of value is")
x = pow(4,3)
print(x)


```

**2. Math Module**

1 math.sqrt()             
2 math.ceil()              
3 math.floor()             
4 math.pi

```python
## 1. math.sqrt()
## calculate square root
## syntax = math.sqrt(number)

import math
x=math.sqrt(64)
print("Print Squareroot of number:")
print(x)

## 2. math.ceil()
## round decimal to upper integer
print("")
import math
print("Print Upper Round of Decimal Number:")
print(math.ceil(1.4))

## 3. math.floor()
## round decimal to down integer
print("")
import math
print("Print Down Round of Decimal Number:")
print(math.floor(1.9))

## 4. math.pi
## it return value of pie
print("")
import math
print("print the value of pie")
print(math.pi)

## 5. pie of circle
print("")
import math
r = 5
area = math.pi * r * r
print(area)
```

**Python _ JSON**

JSON is a syntax for storing and exchanging data                       
JSON is a text, written with javascript object notation                   To use JSON in python we have to import module  without import functions cannot be used

```python
## Data Access
import json
x = '{"name":"John","age":30,"city":"New York"}'
y = json.loads(x)
print("Load data from json")
print(y)
print(y["age"])
print(y["city"])
## loads() transform json into python object

## 2. Python ---> JSON(dumps)
## now convert python dictionary into json
import json
person = {
    "name":"John",
    "age":30
}
result = json.dumps(person)
print("Convert python dictionary into json")
print(result)

print("")
## dict into JSON
import json
print("JSON into dict:")
print(json.dumps({"name":"Ali"}))
print("")
## list into JSON
print("JSON into list")
print(json.dumps(["Apple","Banana"]))
## tuple into tuple
print("")
print("JSON into tuple")
print(json.dumps(("Apple","Banana")))
## Boolean
print("")
print("Boolean")
print(json.dumps(True))
print(json.dumps(False))
print(json.dumps(None))
```

**Python _ RegEx**

It is search pattern       
It is used to find  anything inside string     
The module of RegEx is:   
import re
Syntax = re.search(pattern, string)


```python
## 1. re.search()
import re
txt = "The rain in Spain"
x = re.search("Spain", txt)    
## spain is pattern(pattern the required word) 
# ## txt is string(str from which word is reuqired ) 
# ## re.search will search from string
print("Search for word spain")
print(x)

## if the required word does not in string
print("")
import re
txt = "Python"
x = re.search("Java", txt)
print("The required word does not exists in string")
print(x)

## 2. findall()   it returns all matches
print("")
import re
txt = "The rain in Spain"
x = re.findall("ai",txt)
print("Find from all  string")
print(x)    ## if does not find anything then empty list[] will be shown

## 3. split()   it divide string
print("")
import re
txt = "The rain in Spain"
## /s means space it will break string where it find space
x = re.split("\s", txt)
print("Split the string")
print(x)

## if we want only one split
print("")
import re
txt = "The rain in Spain"
## /s means space it will break string where it find first space
x = re.split("\s", txt, 1)
print("Split the string just once")
print(x)

## 4. sub()    it replaces
print("")
import re
txt = "The rain in Spain"
## where it will find space it will replace that space with 9
x = re.sub("\s","9",txt)
print("Replace space with 9")
print(x)

## replace only 2
print("")
import re
txt = "The rain in Spain"
## where it will find space it will replace that only 2 spaces with 9
x = re.sub("\s","6",txt,2)
print("Replace 2 space with 9")
print(x)

##  5. MetaCharacters These are special symbols

## Dot(.)    it means any one character
print("")
import re
txt="hello"
print("Match the string using dot")
print(re.findall("h.llo",txt))  ## (.) will find e because it is at place of e

## (^)  it represent start of string
print("")
import re
txt="Hello World"
print("Start of string")
print(re.findall("^Hello",txt))

## ($)  it represent end of string
print("")
import re
txt="Hello World"
print("Represent end of string")
print(re.findall("World$",txt))

## (*)  null or many
## (+)  one or many
## (?)  Zero or one
## or(|)    e.g cat or dog
## {}   Specific Quantity  ,  e.g{3} , output: aaa
## []   character set      ,  e.g[a,b,c], a or b or c
print("")
import re
txt="apple"
print("Character set")
print(re.findall("[ae]",txt))

print("")

## 6. Special Characters
## /d  digits 0-9
import re
txt="My age is 20"
print("Show digits")
print(re.findall("\d",txt))


## /D not digits
print("")
import re
txt="Age20"
print("Show No Digits")
print(re.findall("\D",txt))


## /s  space
print("")
import re
txt="Hello World"
print("Show Space")
print(re.findall("\s",txt))

## /S No Space
print("")
import re
txt="A B"
print("Show No Space")
print(re.findall("\S",txt))

## /w  Word Characters (letter,digits, underscore)
print("")
import re
txt="abc123_"
print("Show wordcharacter")
print(re.findall("\w",txt))

## /W   excluding word character
print("")
import re
txt="abc123_"
print("Dont Show word character")
print(re.findall("\W",txt))

## span()   it explain position
print("")
import re
txt="The rain in Spain"
x=re.search("Spain",txt)  ## will  tell start and end position
print("Explain Position")
print(x.span())

## group()   will return input
print("")
import re
txt="The rain in Spain"
x=re.search("Spain",txt)
print("Return input value using group()")
print(x.group())

## string()
print("")
import re
txt="The rain in Spain"
x=re.search("Spain",txt)
print("Return original string using string()")
print(x.string)


## flags  it  is ignore case
print("")
import re
txt="Hello"
print("ignore case using flag")   ## Hello != hello
print(re.findall("hello",txt,re.I))  ## re.I will ignore upper lower case difference
print("")
```

**Python  Try Except**


try  test a block of code for errors.

except     handle the error.                     

else       execute code when there is no error.                     

finally     execute code, regardless of the result of the try and except blocks.

```python
x=5
try:
    print(x)
except:
    print("Error")
print("Hello")

## Types of error
#NameError
#TypeError
#ValueError
#ZeroDivisionError
#FileNotFoundError

## NameError
print("")
x=20
try:
    print(x)
except NameError:
    print("Variable x is not defined")
except:
    print("Some other error")
```

```python
## Name error using try except and else
print(" ")
try:
    x = 10
    y = 5

    print(x+y)
except:
    print("Error")
else:
    print("Addition successful")
```

```python
## finally

print("")
x=15
try:
    print(x)

except:
    print("Error")

finally:
    print("Program Finished")  ## finally block will always be executed 
```
